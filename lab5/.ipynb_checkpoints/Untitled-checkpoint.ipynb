```python
import numpy as np

import matplotlib.pyplot as plt

from scipy.optimize import curve_fit
```

```python
with open('bubbles.txt', 'r') as f:
    bubbles = list(map(lambda x: float(x), f.read().split()))
with open('quick.txt', 'r') as f:
    quick = list(map(lambda x: float(x), f.read().split()))
with open('Bucket.txt', 'r') as f:
    bucket = list(map(lambda x: float(x), f.read().split()))
```

```python
import numpy as np
from scipy.optimize import curve_fit

# Define the linear function
def linear(x, a):
    return a * x

# Define quadratic function
def quadratic(x, a, b):
    return ((x + b) ** 2) * a

# Mock Y data (replace these with actual datasets)
# Ensure no zeros or negative in Y for certain funcs
quick = np.array(quick, dtype=float)
bucket = np.array(bucket, dtype=float)
bubbles = np.array(bubbles, dtype=float)

# Linear fit for quick
X = np.linspace(10, 97000, 1000)[:11]
params, _ = curve_fit(linear, X, quick, p0=[1])  # Initial guess for a
new_X = np.arange(10, int(1e6), 1000)
quick = linear(new_X, *params)

# Linear fit for bucket
params, _ = curve_fit(linear, X, bucket, p0=[1])
bucket = linear(new_X, *params)

# Quadratic fit for bubbles
params, _ = curve_fit(quadratic, X, bubbles, p0=[1, 0])
bubbles = quadratic(new_X, *params)

print("Quick, Bucket, Bubbles fit OK")
```

```python
X = list(range(10, int(1e6), 1000))
y1 = bubbles
y2 = quick
y3 = bucket

# Plot original data and the fitted curve
plt.plot(X, y1, label="Bubble", color="red")
plt.plot(X, y2, label="Quick", color="blue")
plt.plot(X, y3, label="Bucket", color="green")

plt.xscale('log')  # Logarithmic x-axis
plt.yscale('log')  # Logarithmic y-axis

plt.xlabel("X")
plt.ylabel("Y")

plt.legend()

plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Filter the data where 1e4 <= X <= 1e5
X = np.array(list(range(10, int(1e6), 1000)))
mask = (X >= 1e4) & (X <= 1e5)

filtered_y1 = bubbles[mask]
filtered_y2 = quick[mask]
filtered_y3 = bucket[mask]

# Create a boxplot
plt.boxplot([filtered_y1, filtered_y2, filtered_y3], labels=["Bubble", "Quick", "Bucket"], patch_artist=True)

# Add labels and grid
plt.xlabel("Categories")
plt.ylabel("Y Values")
plt.title("Boxplot for Values of Bubble, Quick, and Bucket (1e4 <= X <= 1e5)")

plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
